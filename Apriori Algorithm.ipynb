```python
import pandas as pd
```

```python
dataset  = [["Milk", "Bread", "Wheat Bread"], ["Jam", "Bread", "Butter"], ["Milk", "Jam", "Bread", "Butter"], ["Jam", "Butter"]]
```

```python
dataset
```

```python
pip install mlxtend  
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
te_ary = te.fit(dataset).transform(dataset)
```

```python
df = pd.DataFrame(te_ary,columns=te.columns_)
```

```python
print(df)
```

```python
from mlxtend.frequent_patterns import apriori 
```

```python
apriori(df,min_support = 0.1)
```

```python
apriori(df, min_support=0.1, use_colnames=True)
```

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets['lenght'] = frequent_itemsets['itemsets'].apply(lambda x:len(x))
```

```python
frequent_itemsets
```

```python
frequent_itemsets[(frequent_itemsets['lenght']==3)&(frequent_itemsets['support']>=0.5)]
```

```python
frequent_itemsets[(frequent_itemsets['lenght']==2)]
```
